```python
# Add a new column Revised_Salary, revised salary is 10% more than actual salary
# Then calculate Total additional cost to company
# Which department faces the biggest impact
```

```python
data.fillna({"Department":data.Department.mode()[0],
            "Age":data.Age.mean(),
            "Salary":data.Salary.median()},inplace=True)
```

```python
data
```

```python
data["Revised_Salary"] = data["Salary"]*1.10
```

```python
data
```

```python
data["additional_cost"]= data["Salary"]-data["Revised_Salary"]
```

```python
data
```

```python
data.drop("additional_cost",axis=1,inplace=True)
```

```python
data
```

```python
data["additional_cost"] = data["Revised_Salary"] - data["Salary"]
```

```python
data
```

```python
impact_by_department = data.groupby("Department")["additional_cost"].sum()
```

```python
most_affected_department = impact_by_department.idxmax()
max_impact = impact_by_department.max()

print(f"The department with the biggest impact is: {most_affected_department} (Additional Cost: {max_impact})")
```
